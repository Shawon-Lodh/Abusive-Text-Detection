```
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator
import numpy as np

fig, ax = plt.subplots()

# Example data
people = ('GRU', 'LSTM', 'Random Forest', 'Logistic Regression', 'SVM', 'Naive Bayes', 'KNN')
y_pos = np.arange(len(people))

performance = [95.26,95.21,90.81,92.88,94.19,88.15,91.29]

ax.barh(y_pos, performance, align='center')
ax.set_yticks(y_pos)
ax.set_yticklabels(people)
ax.invert_yaxis()  # labels read top-to-bottom
ax.set_xlabel('Accuracy in (%)')
ax.set_title('Comparison between Traditional ML and Deep Learning at a glance')

for i, v in enumerate(performance):
    plt.text(v + 0.5, i+0.1, str(v)+'%', color='blue', fontweight='bold')

plt.rcParams['figure.figsize'] = (8,5)
plt.show()
```
