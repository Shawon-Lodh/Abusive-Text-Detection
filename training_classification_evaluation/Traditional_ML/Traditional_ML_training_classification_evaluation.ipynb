```
#import Library
import pickle as pkl
import numpy as np
import matplotlib.pyplot as plt
import math
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB, MultinomialNB, BernoulliNB, ComplementNB
from sklearn.neighbors import KNeighborsClassifier
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
```

#Data read

```
#Data read from json file
from google.colab import drive
drive.mount('/content/gdrive')
```

```
train_data_feature = []
test_data_feature = []
train_path = '/content/gdrive/My Drive/Abusive_text_detection/Feature_extraction/Extracted_feature_traditional_ML/Train/train_Feature_extraction_using_'
test_path = '/content/gdrive/My Drive/Abusive_text_detection/Feature_extraction/Extracted_feature_traditional_ML/Test/test_Feature_extraction_using_'
end_path = '_Traditional_ML.pkl'
middle_path = ['trainable_word2vec','glove_twitter_25','glove_twitter_50','glove_twitter_100','glove_twitter_200','fasttext_wiki_news_subwords_300']
for i in middle_path:

  #Load train data from google drive
  with open(train_path+i+end_path,'rb') as f:
    train_data_feature.append(pkl.load(f))

  #Load test data from google drive
  with open(test_path+i+end_path,'rb') as f:
    test_data_feature.append(pkl.load(f))

#Load train level from google drive
with open('/content/gdrive/My Drive/Abusive_text_detection/Feature_extraction/Label/train_label.pkl','rb') as f:
  train_level = pkl.load(f)

#Load test level from google drive
with open('/content/gdrive/My Drive/Abusive_text_detection/Feature_extraction/Label/test_label.pkl','rb') as f:
  test_level = pkl.load(f)
```

#Plot accuracy graph

```
#Plot Accuracy Graph
def plot_graph(model_name,accuracy,xlevel):
  for i in range(len(xlevel)):
    xlevel[i] = xlevel[i].replace("_", "-")
    if xlevel[i] == 'fasttext-wiki-news-subwords-300':
      xlevel[i] = 'fasttext-wiki-news-\nsubwords-300'
      
  ind = np.arange(len(xlevel))  # the x locations for the groups
  width = 0.30        # the width of the bars

  fig = plt.figure()
  ax = fig.add_subplot(111)

  yvals = accuracy
  rects1 = ax.bar(ind, yvals, width, color='b')
  
  low = min(yvals)
  high = max(yvals)
  plt.ylim([math.ceil(low-0.5*(high-low)),  math.ceil(high+0.5*(high-low))])

  
  ax.set_ylabel('Accuracy in %')
  ax.set_xlabel('Word_embedding_type')
  ax.set_xticks(ind)
  ax.set_xticklabels( (xlevel) )
  ax.set_title('Accuracy of '+model_name) 
  
  def autolabel(rects):
      for rect in rects:
          height = rect.get_height()
          ax.annotate('{}%'.format(height),
                      xy=(rect.get_x() + rect.get_width() / 2, height),
                      xytext=(0, 3),  # 3 points vertical offset
                      textcoords="offset points",
                      ha='center', va='bottom')

  autolabel(rects1)
  plt.rcParams['figure.figsize'] = (10,5)
  plt.show()
```

#KNN

```
#Knn trainning,classification and Evaluation
knn_all_accuracy = []
for i in range(len(train_data_feature)):
  knn = KNeighborsClassifier(n_neighbors=9).fit(train_data_feature[i], train_level)
  knn_predictions = knn.predict(test_data_feature[i])
  Accuracy = accuracy_score(test_level, knn_predictions)
  print("when ",middle_path[i].replace("_", "-"),",\nAccuracy : ",round(Accuracy,4),'\nConfusion Matrix : \n',confusion_matrix(test_level, knn_predictions),'\nClassification Report : \n',classification_report(test_level, knn_predictions))
  knn_all_accuracy.append(Accuracy)
```

```
#Knn graph
plot_graph('KNN',knn_all_accuracy,middle_path)
```

#Naive Bayes

```
#Naive Bayes trainning,classification and Evaluation
naive_bayes_all_accuracy = []
for i in range(len(train_data_feature)):
  n_bayes = BernoulliNB().fit(train_data_feature[i], train_level)
  n_bayes_predictions = n_bayes.predict(test_data_feature[i])
  Accuracy = accuracy_score(test_level, n_bayes_predictions)
  print("when ",middle_path[i].replace("_", "-"),",\nAccuracy : ",round(Accuracy,4),'\nConfusion Matrix : \n',confusion_matrix(test_level, n_bayes_predictions),'\nClassification Report : \n',classification_report(test_level, n_bayes_predictions))
  naive_bayes_all_accuracy.append(Accuracy)
```

```
#Naive Bayes graph
plot_graph('Naive Bayes',naive_bayes_all_accuracy,middle_path)
```

#SVM

```
#SVM trainning,classification and Evaluation
svm_all_accuracy = []
for i in range(len(train_data_feature)):
  svm = SVC(kernel='rbf',C=13).fit(train_data_feature[i], train_level)
  svm_predictions = svm.predict(test_data_feature[i])
  Accuracy = accuracy_score(test_level, svm_predictions)
  print("when ",middle_path[i].replace("_", "-"),",\nAccuracy : ",round(Accuracy,4),'\nConfusion Matrix : \n',confusion_matrix(test_level, svm_predictions),'\nClassification Report : \n',classification_report(test_level, svm_predictions))
  svm_all_accuracy.append(Accuracy)
```

```
#Naive Bayes graph
plot_graph('SVM',svm_all_accuracy,middle_path)
```

#Logistic Regression

```
#Logistic Regression trainning,classification and Evaluation
Lr_all_accuracy = []
for i in range(len(train_data_feature)):
  lr = LogisticRegression(C=3,solver='newton-cg',multi_class='auto').fit(train_data_feature[i], train_level)
  lr_predictions = lr.predict(test_data_feature[i])
  Accuracy = accuracy_score(test_level, lr_predictions)
  print("when ",middle_path[i].replace("_", "-"),",\nAccuracy : ",round(Accuracy,4),'\nConfusion Matrix : \n',confusion_matrix(test_level, lr_predictions),'\nClassification Report : \n',classification_report(test_level, lr_predictions))
  Lr_all_accuracy.append(Accuracy)
```

```
#Logistic Regression graph
plot_graph('Logistic Regression',Lr_all_accuracy,middle_path)
```

#Random Forest

```
#Random Forest trainning,classification and Evaluation
ran_for_all_accuracy = []
for i in range(len(train_data_feature)):
  ran_for = RandomForestClassifier(n_estimators=300,max_depth=256,n_jobs=60).fit(train_data_feature[i], train_level)
  ran_for_predictions = ran_for.predict(test_data_feature[i])
  Accuracy = accuracy_score(test_level, ran_for_predictions)
  print("when ",middle_path[i].replace("_", "-"),",\nAccuracy : ",round(Accuracy,4),'\nConfusion Matrix : \n',confusion_matrix(test_level, ran_for_predictions),'\nClassification Report : \n',classification_report(test_level, ran_for_predictions))
  ran_for_all_accuracy.append(Accuracy)
```

```
#Logistic Regression graph
plot_graph('Logistic Regression',ran_for_all_accuracy,middle_path)
```
